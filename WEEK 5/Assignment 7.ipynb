```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic=pd.read_csv("titanic_data.csv")
print(titanic.shape)
titanic.head()
```

# I.Total number of passengers from each class who survived

```python
titanic.groupby(['class'])['survived'].sum().reset_index()
```

# II.Average fare for passengers from each town

```python
titanic.groupby(['embark_town'])['fare'].mean().reset_index()
```

# III. What was the average fare for male vs femalepassengers from each town? (adding more groups/levels)


```python
titanic.groupby(['embark_town','sex'])['fare'].mean().reset_index()
```

# IV. What was the average fare and age for male vs femalepassengers from each town? (Adding morevariables/features)


```python
titanic.groupby(['embark_town','sex'])['fare','age'].mean().reset_index()
```

# V.Get the unique set of ages for all each age category (whocolumn) from each town

```python
titanic.groupby(['embark_town','who'])['age'].apply(set).reset_index()
```

# VI.Get the range (min - max) of ages for each age category(who column) from each town

```python
titanic.groupby(['embark_town','who'])['age'].apply(lambda x:x.max()-x.min()).reset_index()
```

# VII. Get the mean fare by age ratio for each age category(who column) from each town.

```python
titanic.groupby(['embark_town','who']).apply(lambda x: (x['fare']/x['age']).mean())
```

# VIII.Get the mean of fare, AND median of age for each age category (who column) from each town


```python
#Define aggregations as a dictionary
g = {'fare':'mean','age':'median'}
titanic.groupby (['embark_town', 'who']).agg (g).reset_index()
```

# IX.Get the sum & mean of fare, AND median, min and maxof age for each age category (who column) from each town

```python
#Define aggregations as a dictionary
g = {'fare' : ['sum', 'mean'],'age': [ 'median', 'min', 'max' ]}
titanic.groupby(['embark_town','who' ]).agg(g).reset_index ()
```

# x) Create a new column that returns the average fare for the age group (who column) the passenger belongs to.

```python
#Define aggregations as a dictionary
g = {'fare' :lambda x: x.sum(),'age' :lambda x: x.max()}
titanic.groupby(['embark_town', 'who' ]).agg(g).reset_index ()
```
