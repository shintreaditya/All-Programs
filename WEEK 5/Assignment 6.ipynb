# ii) Illustrate the usage of join to combine data frames                            

```python
import pandas as pd
df1 = pd. DataFrame ({ 'key': ['KO', 'K1', 'K5', 'K3', 'K4', 'K2'],'A': ['A0', 'A1','A5','A3','A4','A2']})
df2= pd. DataFrame ({ 'key': ['KO', 'K1', 'K2'],'B': ['BO', 'B1', 'B2']})
```

# Case 1:Join on indexes

```python
#default nature of pandas join is left outer join
df1.join(df2, lsuffix=' _a', rsuffix=' _b')
```

# Create two dataframes with different index values.

```python
df1 = pd. DataFrame ({ 'key': ['KO', 'K1', 'K5', 'K3', 'K4', 'K2'],'A': ['A0', 'A1','A5','A3','A4','A2']},index= [0,1,2,3,4,5])
df2 = pd. DataFrame ({ 'key': ['KO', 'K1', 'K2'],'B': ['B0', 'B1', 'B2']},index= [2,0,1])
df1.join (df2,lsuffix=' _l', rsuffix=' _r')
#df1 is left DF and df2 is right DF
```

```python
#inner join
df1.join (df2,lsuffix=' _l', rsuffix=' _r',how='inner')############################
```

```python
#outer join
df1.join (df2,lsuffix='_l', rsuffix='_r', how='outer')
```

# Case 2.Join on columns


```python
import pandas as pd
df1=pd.DataFrame({'key1':['K0','K1','K5','K3','K4','K2'],'A':['A0','A1','A5','A3','A4','A2']})
df2=pd.DataFrame({'key2':['K0','K1','K2'],'B':['B0','B1','B2']})
df1=pd.DataFrame({'key1':['K0','K1','K5','K3','K4','K2'],'A':['A0','A1','A5','A3','A4','A2']})
df2=pd.DataFrame({'key2':['K0','K1','K2'],'B':['B0','B1','B2']},index=[6,7,8])
```

```python
#join on data frame column
df1.set_index('key1').join(df2.set_index ('key2'))
```

# iii.Illustrate the use of concat to combine dataframes

## Case 1. concat data frames on axis=0, default operation


```python
import pandas as pd
m1 =pd. DataFrame ({'Name': [ 'Alex', 'Amy', 'Allen', 'Alice', 'Ayoung' ],'subject_id':['sub1','sub2','sub4','sub6','sub5'],'Marks_scored': [98, 90, 87, 69, 781]},index= [1, 2, 3, 4, 5])
m2 =pd. DataFrame ({'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],'subject_id' :['sub2','sub4','sub3','sub6','sub5'],'Marks_scored': [89, 80, 79,97,88]},index= [ 4, 5, 6, 7, 8])
pd.concat([m1, m2])
```

## Case 2. concat operation on axis=1, horizontal operation

```python

#axis=1 works as join operation
pd.concat([m1, m2], axis=1) #######################################################
```

## Case 3. concat unequal shape data frames


```python
df1 = pd. DataFrame ({ 'A': [1,2,3], 'B': [1,2,3]})
df2= pd. DataFrame ({ 'A': [4, 5] })
```

```python
df = pd.concat([df1, df2],ignore_index=True)
df
```
