# 9.Obtain distribution of data of various columns

```python
#Obtaining the distribution of data of various columns
fig,axes=plt.subplots(2,2,figsize=(16,9))
axes[0,0].set_title('Distribution of sepal width')
axes[0,0].hist(df['SepalWidthCm'],bins=5)
axes[0,1].set_title('Distribution of sepal length')
axes[0,1].hist(df['SepalLengthCm'],bins=7)
axes[1,0].set_title('Distribution of Petal width')
axes[1,0].hist(df['PetalWidthCm'],bins=5)
axes[1,1].set_title('Distribution of petal length')
axes[1,1].hist(df['PetalLengthCm'],bins=6)
```

# 10.Perform univariate analysis

```python
#Petal Width
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"PetalWidthCm").add_legend()
```

```python
#Petal Length
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"PetalLengthCm").add_legend()
```

```python
#Sepal Length
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"SepalLengthCm").add_legend()
```

```python
#Sepal Width
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"SepalWidthCm").add_legend()
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,9))
sns.boxplot(y='PetalWidthCm',x="Species",data=df,orient='v', ax=axes[0,0])
sns.boxplot(y='PetalLengthCm',x="Species",data=df, ax=axes[0,1])
sns.boxplot(y='SepalLengthCm',x="Species",data=df, ax=axes[1,0])
sns.boxplot(y='SepalWidthCm',x="Species",data=df, ax=axes[1,1])
plt.show()


```

```python
fig,axes = plt.subplots(2,2,figsize=(16,9))
sns.violinplot(y='PetalWidthCm',x="Species",data=df,orient='v', ax=axes[0,0])
sns.violinplot(y='PetalLengthCm',x="Species",data=df, ax=axes[0,1])
sns.violinplot(y='SepalLengthCm',x="Species",data=df, ax=axes[1,0])
sns.violinplot(y='SepalWidthCm',x="Species",data=df, ax=axes[1,1])
plt.show()
```
